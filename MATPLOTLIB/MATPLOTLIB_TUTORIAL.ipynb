```python
# Matplotlib is a Python package used for 2D graphics
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3])
y = np.array([4,5,6])
```

```python
plt.plot(x,y)

plt.title('Basic Plot')

plt.xlabel('X-axis')

plt.ylabel('y-axis')

plt.show()
```

```python
from matplotlib import style
```

```python
a = [7,8,9]
b = [10,11,12]
```

```python
style.use('ggplot')
```

```python
plt.plot(x,y,'red',label='Line 1',linewidth=5)
plt.plot(a,b,'yellow',label='Line 2',linewidth=5)

plt.legend(loc='lower right')

plt.grid(True,color='green')

plt.show()
```

```python
# Bar Graph
```

```python
plt.bar(x,y,color='c',label='First Sales')
plt.bar(a,b,color='green',label='Second Sales')

plt.legend()

plt.show()
```

```python
# Histogram
```

```python
population_ages = [22,55,62,45,21,22,34,42,42,4,99,102,110,120,121,122,130,111,115,112,80,75,65,54,44,43,42,48]

bins = list(range(0,140,10))
```

```python
plt.hist(population_ages,bins,histtype='bar',rwidth=0.8)

plt.legend()

plt.show()
```

```python
# Scatter plot
```

```python
plt.scatter(x,y,label='scatter',marker='x',color='k')

plt.legend()

plt.grid(False)

plt.show()
```

```python
# Stack Plot
```

```python
days = [1,2,3,4,5]

sleep = [7,8,6,11,7]
play = [8,5,7,8,13]
eat = [2,3,4,3,2]
work = [7,8,7,2,2]

plt.plot([],[],color='green',label='sleep',linewidth=5)
plt.plot([],[],color='red',label='play',linewidth=5)
plt.plot([],[],color='yellow',label='eat',linewidth=5)
plt.plot([],[],color='cyan',label='work',linewidth=5)

plt.stackplot(days,sleep,eat,work,play,colors=['green','red','yellow','cyan'])

plt.legend()

plt.show()
```

```python
# Pie Chart

slices = [7,2,2,13]

activities = ['sleep','eat','work','play']

cols = ['m','c','r','k']

plt.pie(slices,labels=activities,colors=cols,startangle=90,shadow=True,explode=(0,0.41,0,0.1),autopct='%1.1f%%')

plt.show()
```

```python
# Working with multiple plots

def f(t):
    return np.exp(-1*t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.subplot(211)

plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))

plt.show()
```
